```python
# 1) Create trimmed CSVs using pandas
import pandas as pd

df = pd.read_csv("data/bank-full.csv", sep=";")

# examples: keep first N rows
for n in [500, 1000, 5000, 10000, 20000, 45211]:
    df.iloc[:n].to_csv(f"data/bank_{n}.csv", sep=";", index=False)
```

```python
# 2) Random sampling - Scientifically Better

import pandas as pd

df = pd.read_csv("bank-full.csv", sep=";")

for n in [500, 1000, 5000, 10000, 20000]:
    df.sample(n=n, random_state=42).to_csv(f"bank_sample_{n}.csv", sep=";", index=False)

```
